# **Three sorting string algorithms**

```python
# Quicksort with 2 parteion 
def qsort(mylist):
    """
      quick_sort: One of the most used sorting algorithm. 
    """
    
    if mylist == []:# Base case: single element list is sorted.
         return mylist
    pivot = mylist[0] # first element in the array is the pivot

    left=qsort([i for i in mylist[1:] if i < pivot])         #left will store therecursive call of all The lower bound of the array exclude the first element   .

    right=qsort([i for i in mylist[1:] if i >= pivot])         # right will store the recursive call of all The equial or higher bound of the array exclude the first element

    return left + [pivot] + right # return left elements + pivot + right elements
    #will repate the same process and each time the pivot will change  until  all elements in array sorted left side smaller the  pivot and right side bigger then the pivot
#######################################################################################################################################################################################################
# Quicksort with 3 parteion 
def three_qsort(list):
    """
  Three way quick sort : best version of quick sort to handle few equal elements in array 
    """
    if len(list) <= 1: # base case if the list elements less then or equal to 1 return the list as the result
         return list
    
    pivot = list[0] # get the first element as pivot
    equal_elements  = [x for x in list if x == pivot] # a list containing all elements equal to the pivot
    smaller_elements = [x for x in list if x <  pivot]# a list containing all elements less then the pivot
    bigger_elements = [x for x in list if x >  pivot]# a list containing all elements bigger then  the pivot
    return three_qsort(smaller_elements) + equal_elements + three_qsort(bigger_elements) # return recarsive call  for smaller elements + equal elements in middle + recarsive call  for bigger elements
    #will repate the same process and each time the pivot will change  until  all elements in array sorted left side contians elements  smaller the  pivot ,the middle with equal elements and  right side contians elemens bigger then the pivot
#######################################################################################################################################################################################################
# Merge sort algo
def merge(leftside, rightside):
    if len(leftside) == 0: # base case if the array is empty then return the second array as the result
        return rightside
    if len(rightside) == 0: # base case if the array right-side is empty then return the left-side of the array as the result
        return leftside

    result = []
    index_leftside =0
    index_rightside = 0
    array_length=len(leftside) + len(rightside)
    while len(result) < array_length : # repeat until the length of result list not less then the array length
        if leftside[index_leftside] <= rightside[index_rightside]: # start with first index in leftside array and check if its less then or equal to to the first index in rightside
            result.append(leftside[index_leftside]) # add the left side element to the array 
            index_leftside += 1 # increase the index by 1 each time the if condation repeated to move to next index 
        else: 
            result.append(rightside[index_rightside])  # otherwise add the right side as the lower element
            index_rightside += 1# increase the index by 1 each time the else condation repeated to move to next index 

        if index_rightside == len(rightside):# If the index right equal to the length of rightside array means that the index reach the  the end of the array 
            result += leftside[index_leftside:] # add the remaning indexs after leftside index  
            break

        if index_leftside == len(leftside):# If the index left equal to the length of leftside array means that the index reach the  the end of the array 
            result += rightside[index_rightside:] # add the remaning indexs after leftside index  
            break

    return result # return result list 

def merge_sort(list):
    
    if len(list) < 2:# base case If the input array contains fewer than two elements,
        return list    # then return it as the result        
    return merge(merge_sort(list[:len(list)//2]),merge_sort(list[len(list)//2:]))  #recursively call for left and right arrays to be sorted and then merge them
```

**Function to extract the data from a given URL**

```python
from urllib.request import urlopen
def extract_txt_data(url="https://www-cs-faculty.stanford.edu/~knuth/sgb-words.txt"):

	file = urlopen(url)
	text=[]
	for line in file:
			decoded_line = line[:-1].decode("utf-8")
			text.append(decoded_line)
	return text
```
